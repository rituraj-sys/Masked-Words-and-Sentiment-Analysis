```python
from transformers import BertTokenizer, BertModel
```

```python
tokenizer = BertTokenizer.from_pretrained("google-bert/bert-large-uncased-whole-word-masking")
```

```python
model = BertModel.from_pretrained("google-bert/bert-large-uncased-whole-word-masking")
```

```python
input = "hello, how are you, i am good"
encoded = tokenizer(input, return_tensors="pt")
```

```python
output = model(**encoded)
output
```

```python
decoded_text = tokenizer.decode(encoded['input_ids'][0])

print("Decoded Text:", decoded_text)
```

```python
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-large-uncased-whole-word-masking')
```

```python
net_word = unmasker("Hello I'm a [MASK] engineer.")
```

***Predicting Masked Words***

```python
# Predicting possible masked words
for i in net_word:
  print(i)
```

**Sentiment Analysis pipeline**

```python
# Loading a sentiment-analysis pipeline
sentiment_analyzer = pipeline('sentiment-analysis')
```

```python
# Analyze sentiment
result = sentiment_analyzer("you play too bad")
print(result)
```
